```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def generate_colors(num_colors):
    return plt.cm.gist_rainbow(np.linspace(0, 1, num_colors))

# Number of circles in the ring
num_circles = 40

# Generate colors for the circles
colors = generate_colors(num_circles)


```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Plot filled circles in a ring
radius = 1.5
theta = np.linspace(0, 2*np.pi, num_circles, endpoint=False)

for i in range(num_circles):
    x = radius * np.cos(theta[i])
    y = radius * np.sin(theta[i])
    
    circle = plt.Circle((x, y), 0.11, color=colors[i])
    ax.add_patch(circle)

# Set aspect ratio to equal for a circular plot
ax.set_aspect('equal')

# Set axis limits
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

# Hide the axes
ax.set_axis_off()
plt.savefig('sensory_network.pdf', dpi=25)

# Show the plot
plt.show()
```

```python
num_circles_line = 20
# Create a figure and axis
fig, ax = plt.subplots()
# Plot a line of filled circles in black next to the ring
for i in range(num_circles_line):
    x_line = 0  # Adjust the x-coordinate as needed
    y_line = 0.2 * i - 0.2 * (num_circles_line - 1) / 2
    
    circle_line = plt.Circle((x_line, y_line), 0.09, color='black')
    ax.add_patch(circle_line)

# Set aspect ratio to equal for a circular plot
ax.set_aspect('equal')

# Set axis limits
ax.set_xlim(-3, 3)
ax.set_ylim(-2, 2)

# Hide the axes
ax.set_axis_off()
plt.savefig('sensory_network.pdf', dpi=25)
plt.show()
```

```python
# Number of circles in the ring
num_circles_ring = 40
num_circles_line = 20

# Generate colors for the circles in the ring
colors_ring = generate_colors(num_circles_ring)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot filled circles in a ring
radius = 0.8
theta_ring = np.linspace(0, 2*np.pi, num_circles_ring, endpoint=False)


# Plot a line of filled circles in black next to the ring
for i in range(num_circles_line):
    x_line = 2.5  # Adjust the x-coordinate as needed
    y_line = 0.2 * i - 0.2 * (num_circles_line - 1) / 2
    
    circle_line = plt.Circle((x_line, y_line), 0.09, color='black')
    ax.add_patch(circle_line)

# Draw 800 lines connecting the circles in the ring to the black circles
for j in range(num_circles_ring):
    for i in range(num_circles_line):
        x_ring = radius * np.cos(theta_ring[j])
        y_ring = radius * np.sin(theta_ring[j])
        
        x_line = 2.5
        y_line = 0.2 * i - 0.2 * (num_circles_line - 1) / 2
        
        # Randomly choose green or red for the line color
        line_color = 'green' if np.random.rand() < 0.5 else 'red'

        # Only plot for 35% of the neurons
        if np.random.rand() < 0.35:
            ax.plot([x_ring, x_line], [y_ring, y_line], color=line_color, lw=0.01)


for i in range(num_circles_ring):
    x = radius * np.cos(theta_ring[i])
    y = radius * np.sin(theta_ring[i])
    
    circle_ring = plt.Circle((x, y), 0.05, color=colors_ring[i])
    ax.add_patch(circle_ring)

# Set aspect ratio to equal for a circular plot
ax.set_aspect('equal')

# Set aspect ratio to equal for a circular plot
ax.set_aspect('equal')

# Set axis limits
ax.set_xlim(-3, 3)
ax.set_ylim(-2, 2)

# Hide the axes
ax.set_axis_off()

# Save the figure
plt.savefig('network_architecture.pdf', dpi=25)

# Show the plot
plt.show()
```
